####  2) Melhorando a interação com o API
Observe dois URLs de consultas diferentes, por exemplo o URL utilizado em aula, e o URL feito no exercício anterior. Compare-os e observe as diferenças.

1. Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado (```'Estado'```ou ```Município```), e que devolve os dados da consulta no formato *DataFrame*.
2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com *status* em 'Arquivado por decurso de prazo' estão registradas?
3. Qual é o município da Bahia com mais solicitações deferidas?
4. Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

1) Seu código aqui

https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=SP&tipo_interessado=Estado
    
https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=RJ&tipo_interessado=Munic%C3%ADpio

No link, os parâmetros estão na última parte. O primeiro parâmetro que é o Estado selecionado está na variável 'uf', portanto, no primeiro conta SP e no segundo RJ, conforme substituição sugerida no exercício. Já o segundo parâmetro é a seleção das opções entre 'Estado' e 'Município, o que é inserido na variável 'tipo_interessado'. Ambas as variáveis são combinadas e utilizadas para geral a URL da consulta.

````python
# 1) Seu código aqui
def consulta_pvl(uf,tipo_interessado):
    url_fun = f'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={uf}&tipo_interessado={tipo_interessado}'
    r_fun = requests.get(url_fun)
    data_json_fun = r_fun.json()
    return pd.DataFrame(data_json_fun['items']) 
    
  
````

````python
# 2) Seu código aqui

resultado = consulta_pvl('MG','Município') 
status = resultado[resultado['status'] == 'Arquivado por decurso de prazo']
status['status'].value_counts()
````

````python
# 3) Seu código aqui
resultado = consulta_pvl('BA','Município')
status = resultado[resultado['status'] == 'Deferido']
status['interessado'].value_counts()
#Luís Eduardo Magalhães (16)
````

````python
# 4) Seu código aqui
resultado = consulta_pvl('BA','Estado')
resultado.to_csv('dataframe_pvl_estado.csv')
````
