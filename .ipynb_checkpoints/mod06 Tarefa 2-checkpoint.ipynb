```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sinasc_ro_2019 = pd.read_csv('SINASC_RO_2019.csv')
print ("Número de linhas e colunas da tabela: {}".format(sinasc_ro_2019.shape))

sinasc_ro_2019.head()
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
sinasc_ro_2019[['IDADEMAE']].boxplot()
```

```python
sns.boxplot(y='IDADEMAE',data=sinasc_ro_2019)
```

####  2) Faça um boxplot do peso do bebê

```python
sinasc_ro_2019[['PESO']].boxplot()
```

```python
sns.boxplot(y='PESO',data=sinasc_ro_2019)
```

####  3) Faça um histograma da idade da mãe

```python
sinasc_ro_2019['IDADEMAE'].plot.hist(bins = 20)
```

```python
sns.displot(data=sinasc_ro_2019,
            alpha = 0.5,
            x = 'IDADEMAE',
            bins = 20,
            element = 'step',
            kde = True
           )
```

#### 4) Faça um histograma do peso do bebê

```python
sinasc_ro_2019['PESO'].plot.hist(bins = 50)
```

```python
sns.displot(data=sinasc_ro_2019,
            alpha = 0.5,
            x = 'PESO',
            bins = 50,
            element = 'step',
            kde = True
           )
```
