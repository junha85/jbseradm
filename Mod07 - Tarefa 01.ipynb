#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
titanic_sem_na[['alive', 'survived','sibsp']].groupby(['alive', 'survived']).count()
````

````python
titanic_sem_na = titanic_sem_na.drop('alive',axis=1)
````

````python
titanic_sem_na[['class', 'pclass','sibsp']].groupby(['class', 'pclass']).count()
````

````python
titanic_sem_na = titanic_sem_na.drop('pclass',axis=1)
titanic_sem_na.head(5)
````

````python
titanic_dummies = pd.get_dummies(titanic_sem_na, columns=['class','who'], drop_first=True)
titanic_dummies.head(5)
````

````python
titanic_dummies.sex = titanic_dummies.sex.map({'female': 1, 'male': 0})
titanic_dummies.sex.unique()
````
